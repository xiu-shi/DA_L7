```python
import numpy as np
import pandas as pd
import statistics as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Freshwater withdrawals from 1993 to 2020 
#years = list(range(1993, 2021)) 

#This is a time serious, continuous data ordered by year

# unit of measure of Annual Freshwater Withdrawals, total (billion cubic meters )
# since the dataset isn't too big, I've created a list by order of the table 'Annual freshwate withdrawals, total column'

water_withdrawals = [18.5502, 21.1102, 19.8502, 22.1602, 21.9282, 21.8222, 21.7162, 21.65924, 21.534, 20.8338,
                     20.1336, 19.4334, 18.7332, 17.55665, 16.3801, 15.20355, 14.027, 13.175, 13.215, 15.868,
                     19.148, 18.463, 9.22, 10.623, 12.05864137, 12.33, 10.53, 8.64]


#Create a DataFrame
water_withdrawals_df = pd.DataFrame(water_withdrawals, columns=['water_withdrawals'])
```

```python
# Calculate mean, median, mode, max, min, range and standard deviation

#mean - The average water withdrawal from year 1993 to 2020 
water_withdrawals_mean = round(water_withdrawals_df['water_withdrawals'].mean(),3)

#median The middle value of withdrawal from year 1993 to 2020 omitted from highest and lowest data
water_withdrawals_median = round(water_withdrawals_df['water_withdrawals'].median(),3)

#mode - The most frequent appeared value/s
water_withdrawals_mode = round(water_withdrawals_df['water_withdrawals'].mode(),3)


#max - The highest value
water_withdrawals_max = round(water_withdrawals_df['water_withdrawals'].max(),3)

#min - The smallest value
water_withdrawals_min = round(water_withdrawals_df['water_withdrawals'].min(),3)

#range - The span of the highest and lowest value, 13.52%
water_withdrawals_range = water_withdrawals_max - water_withdrawals_min

#variance
water_withdrawals_variance = round(water_withdrawals_df['water_withdrawals'].var(),3)

#standard deviation - The indicator of how spead out the data is around the mean. STD result is: 4.261%
water_withdrawals_std = round(np.std(water_withdrawals_df['water_withdrawals']),3)
```

```python
print(f"Mean: {water_withdrawals_mean}%")
print(f"Median: {water_withdrawals_median}%")
#Mode: there isn't repeated values in this dataset
print(f"Mode: \n{water_withdrawals_mode}%")

print(f"Max: {water_withdrawals_max}%")
print(f"Min: {water_withdrawals_min}%")
print(f"Range: {water_withdrawals_range}%")
print(f"Variance: {water_withdrawals_variance}%")
print(f"Standard Deviations: {water_withdrawals_std}%")
```
