```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import numpy as np
```

```python
# Descriptive Statistics 1 - Residential Water Usage in Australia in 2021
# Create a DataFrame to represent Residential Water Usage in Australia in 2021 among 6 residential areas, total 100%

residential_data = {
    'Usage': ['Bathroom', 'Toilet', 'Washing Clothes', 'Kitchen', 'Garden'],
    'AUS Residential Water Consumption (%)': [28, 15, 20, 10, 27]
}
# Convert data to a DataFrame
df = pd.DataFrame(residential_data)
residential_df = pd.DataFrame(residential_data)
```

```python
# Calculate mean, median, mode, max, min, range and std for AUS Residential Water Consumption
residential_mean = residential_df['AUS Residential Water Consumption (%)'].mean()
residential_median = residential_df['AUS Residential Water Consumption (%)'].median()
residential_mode = residential_df['AUS Residential Water Consumption (%)'].mode()

residential_max = residential_df['AUS Residential Water Consumption (%)'].max()
residential_min = residential_df['AUS Residential Water Consumption (%)'].min()
residential_range = residential_max - residential_min
residential_std = np.std(residential_df['AUS Residential Water Consumption (%)'])
```

```python
print(f"Mean: {residential_mean}%")
print(f"Median: {residential_median}%")
print(f"Mode: \n{residential_mode}%")

print(f"Max: {residential_max}%")
print(f"Min: {residential_min}%")
print(f"Range: {residential_range}%")
print(f"Standard Deviations: {residential_std}%")
```

```python
# Plotting Histogram

plt.figure(figsize=(6, 3))
sns.histplot(df['AUS Residential Water Consumption (%)'])
plt.title('Residential Water Consumption')
plt.xlabel('Water Consumption by %')
plt.ylabel('Frequency')
plt.show()
# it's not a bell curve shape, indicating it's not a normally distributed data
```

```python
# Plotting a boxplot to check for Symmetry

plt.figure(figsize=(6, 3))
sns.boxplot(x=df['AUS Residential Water Consumption (%)'])
plt.title('Boxplot of Residential Water Consumption')
plt.xlabel('Percentage of Water Consumption')
plt.show()
```

```python
#Skew value (https://community.gooddata.com/metrics-and-maql-kb-articles-43/normality-testing-skewness-and-kurtosis-241)
#If skewness is less than -1 or greater than 1, the distribution is highly skewed.
#If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
#
#If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.

skewness = df['AUS Residential Water Consumption (%)'].skew()
skewness
```
