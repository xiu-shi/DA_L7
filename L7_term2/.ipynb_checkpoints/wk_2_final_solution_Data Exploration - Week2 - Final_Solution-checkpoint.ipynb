### Drop all incomplete cases

```python
#copy the dataset
df_ts_treated_Drop = df_ts.copy()
## drop all na values
df_ts_treated_Drop.dropna(inplace=True)
##describe the dataset
df_ts_treated_Drop.describe()
```

### Impute with the median

```python
#copy the dataset
df_ts_treated_Median = df_ts.copy()

#replace NA by median
df_ts_treated_Median[['popularity']] = df_ts_treated_Median[['popularity']].replace(np.nan,
df_ts_treated_Median[['popularity']].median())
```

```python
# check if there is any null value 
df_ts_treated_Median[['popularity']].isnull().sum()
```

```python
# describe the new dataset 
df_ts_treated_Median.describe()
```

```python
## describe the previous and compare them 
df_ts.describe()
```

#### Input using SimpleImputer

```python
### https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html
```

```python
## copy the dataset
df_ts_treated_SI = df_ts.copy()
## perfomr simple imputer 
imputer = SimpleImputer(missing_values=np.nan, strategy="median")

#### here we are applying to all columns only for the objetive of showing you the function accepts multiple columns. (We already know  for "popularity" is an
### acceptable method. but for all other columns, more investigation would be required. 
df_ts_treated_SI[measure_cols] = imputer.fit_transform(df_ts_treated_SI[measure_cols])
```

```python
##check null values
df_ts_treated_SI.isnull().sum()
```

```python
## describe the treated dataset 
df_ts_treated_SI.describe()
```

### Visual data exploration methods - after data treatment 

#### Combining both graphs

```python

## plot the treated and the original dataset popularity 

plt.figure(figsize=(20, 10))
sns.set(font_scale=2)
sns.kdeplot(
    data=df_ts_treated_SI,
    x="popularity",
    bw_adjust = 1
)
sns.kdeplot(
    data=df_ts, 
    x="popularity", 
    color="red", 
    bw_adjust = 1
)
```

```python

# perform a grid with all columns treated 
axes =  figsize=(30,15)
cols = measure_cols  # create a list of dataframe columns to use

fig, axes = plt.subplots(ncols=len(cols), figsize=(30,15))

for ax, col in zip(axes, cols):
    sns.kdeplot(data=df_ts_treated_SI[col],ax = ax)
    sns.kdeplot(data=df_ts[col],ax = ax)
   
plt.show()
```

#### Correlation matrix

```python
# check potential relationships using the treated data 
sns.set(font_scale=1)
ax = sns.heatmap(
    df_ts_treated_SI[measure_cols].corr(),
    vmin=-1, 
    vmax=1, 
    annot=True
)
ax.figure.subplots_adjust(left=0.2, bottom=0.2)
ax
```
