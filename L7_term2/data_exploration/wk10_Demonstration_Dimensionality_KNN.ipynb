```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
import seaborn as sns
```

```python
cl1_df = np.random.normal(0,1,(20,1))
cl2_df = np.random.normal(3,1,(20,1))
```

```python
data = np.concatenate([cl1_df, cl2_df], axis = 0)
```

```python
distances = cdist(data,data,metric='euclidean').flatten()
distances
```

```python


plt.scatter(data[:20],[0]*20, s=10, color='r')
plt.scatter(data[20:],[0]*20, s=10, color='b')
plt.title('1 dimension')
```

```python
sns.histplot(distances, kde=True)
```

```python
#### working with 2 dimensional data 
cl1_df = np.random.normal(0,1,(20,2))
cl2_df = np.random.normal(3,1,(20,2))
data = np.concatenate([cl1_df, cl2_df], axis = 0)
```

```python
plt.scatter(data[:20,0],data[:20,1], s=10, color='r')
plt.scatter(data[20:,0], data[:20,1], s=10, color='b')
plt.title('2 Dimensions')
```

```python
distances = cdist(data,data,metric='euclidean').flatten()
sns.histplot(distances, kde=True)
plt.title('2 Dimensions')
```

```python
#### working with 3 dimensional data 
cl1_df = np.random.normal(0,1,(20,3))
cl2_df = np.random.normal(3,1,(20,3))
data = np.concatenate([cl1_df, cl2_df], axis = 0)
```

```python
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(data[:20,0], data[:20,1], data[:20,2], color = "red")
ax.scatter3D(data[20:,0], data[20:,1], data[20:,2], color = "blue")
plt.title('3 Dimensions')
```

```python
distances = cdist(data,data,metric='euclidean').flatten()

sns.histplot(distances, kde=True)
plt.title('3 Dimensions')
```
