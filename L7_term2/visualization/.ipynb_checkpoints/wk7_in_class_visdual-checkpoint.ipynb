```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/Users/eupirate/Documents/DA_L7/L7_term2/visualization/wk7_in_class_Cereals_Rating.csv")
dataset.head()
```

```python
# we create a bar chart with the top 10 cereals by calories

plt.figure(figsize=(10,5))
bar_plot=sns.barplot(x="calories",y="name",data=dataset.sort_values("calories",ascending=False).head(10))
bar_plot.set_title("Top 10 Cereals by Calories")

bar_plot.set_xlabel("Calories")
bar_plot.set_ylabel("Cereal Name")
plt.show()
```

```python
# we create a bar chart with the top 10 cereals by calories - diff colours

plt.figure(figsize=(10,6))
bar_plot=sns.barplot(x="calories",y="name",data=dataset.sort_values("calories",ascending=False).head(10),palette="coolwarm")
bar_plot.set_title("Top 10 Cereals by Calories")

bar_plot.set_xlabel("Calories")
bar_plot.set_ylabel("Cereal Name")
plt.show()
```

```python
# Change background colour
# BG Colour options: white, dark, whitegrid, darkgrid, ticks

# set background 

sns.set(style = 'darkgrid')
plt.figure(figsize = (10,6))
bar_plot = sns.barplot(x = 'calories', y='name', data = dataset.sort_values('calories',ascending = False).head(10), 
                       palette = 'viridis')
bar_plot.set_title('Top 10 Cereals by Calories')
bar_plot.set_xlabel('Calories')
bar_plot.set_ylabel('Cereal Name')
plt.show()
```

```python
# we create a pie chart showing distribution of the cereal by manufacturer

mfr_distribution=dataset["mfr"].value_counts()
mfr_distribution
```

```python
# Define the colour
palette="pastel"

# Create chart
plt.figure(figsize=(6,6))
plt.pie(mfr_distribution,labels=mfr_distribution.index,autopct="%1.1f%%",
        colors=sns.color_palette(palette,len(mfr_distribution)))
plt.title("Cereal Distribution by Manufacturer")
plt.show()
```

```python
# Prepare data -> compare mfr, type and shelf

mfr_distribution=dataset["mfr"].value_counts()
type_distribution=dataset["type"].value_counts()
type_distribution

```

```python
shelf_distribution=dataset["shelf (Cff)"].value_counts()
shelf_distribution
```

```python
# create diff charts in one chart

fig, ax = plt.subplots(1, 3, figsize=(18, 8))

# Pie chart for mfr_distribution
ax[0].pie(mfr_distribution, labels=mfr_distribution.index, autopct="%1.1f%%", startangle=140,
          colors=sns.color_palette("bright"))
ax[0].set_title("Cereal Distribution by Manufacturer")

# Pie chart for type distribution
ax[1].pie(type_distribution, labels=type_distribution.index, autopct="%1.1f%%", startangle=140,
          colors=sns.color_palette("pastel"))
ax[1].set_title("Cereal Distribution by Type")

# Pie chart for shelf distribution
ax[2].pie(shelf_distribution, labels=shelf_distribution.index, autopct="%1.1f%%", startangle=140,
          colors=sns.color_palette("dark"))
ax[2].set_title("Cereal Distribution by Shelf")
```

```python
# create barchart in the same plot

fig, ax = plt.subplots(1, 3, figsize=(18, 8))

# create bar chart for mfr
sns.barplot(x=mfr_distribution.index,y=mfr_distribution.values,ax=ax[0],palette="bright")
ax[0].set_title("Cereal Distribution by Manufacturer")
ax[0].set_xlabel("Manufacturer")
ax[0].set_ylabel("Distribution Count")

# create barchart for the type

sns.barplot(x=type_distribution.index,y=type_distribution.values,ax=ax[1],palette="pastel")
ax[1].set_title("Cereal Distribution by Manufacturer")
ax[1].set_xlabel("Type")
ax[1].set_ylabel("Distribution Count")

# create bar chart for the shelf

sns.barplot(x=shelf_distribution.index,y=shelf_distribution.values,ax=ax[2],palette="pastel")
ax[2].set_title("Cereal Distribution by Manufacturer")
ax[2].set_xlabel("Shelf")
ax[2].set_ylabel("Distribution Count")


plt.show()
```

```python
# add percentages in the plot


# we prepare the data

mfr_distribution=dataset["mfr"].value_counts()
total=mfr_distribution.sum()
percentage=(mfr_distribution/total)*100
```

```python
# create bar chart

plt.figure(figsize=(10,6))
barplot = sns.barplot(x=mfr_distribution.index, y =mfr_distribution.values, palette = 'bright')

#Add the percentage tp the graph

for index, value in enumerate(mfr_distribution.values):
    plt.text(index, value, f'{percentage[index]:.1f}%', ha = 'center', va ='bottom')
    
plt.title('Cereal Distribution by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Distribution Count')
plt.show()
```
