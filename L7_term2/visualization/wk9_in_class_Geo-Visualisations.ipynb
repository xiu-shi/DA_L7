```python
#We want to locate cities in Ireland

#Dublin, Cork, Galway, Limerick, Waterford

!pip install folium
```

```python
import folium
```

```python
locations = [
    ['Dublin',53.349805, -6.260310],
    ['Cork', 51.898514, -8.475604],
    ['Galway',53.270668, -9.056791],
    ['Limerick',52.663837, -8.626734],
    ['Waterford', 52.259320, -7.110070]
]
```

```python
#Create a map that shows Ireland and the cities

ireland_map = folium.Map(location=[53.1424, -7.6921], zoom_start=6)
```

```python
#We add the cities in the map

for town in locations:
    folium.Marker(
    location = [town[1], town[2]], 
    popup=f"{town[0]}",
    icon= folium.Icon(color='green', icon = 'info-sign'),
    ).add_to(ireland_map)

ireland_map
```

```python
###WORLD data

country_data = px.data.gapminder()
country_data.head()
```

```python
map_fig = px.scatter_geo(country_data, locations = 'iso_alpha', projection = 'natural earth', color = 'continent',
                        hover_name = 'country',
                        hover_data = ['lifeExp', 'pop', 'year'])
map_fig.show()
```

```python
map_fig = px.scatter_geo(country_data, locations = 'iso_alpha', projection = 'orthographic', color = 'continent',
                        hover_name = 'country',
                        hover_data = ['lifeExp', 'pop', 'year'])
map_fig.show()
```
