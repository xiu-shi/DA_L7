```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("/Users/eupirate/Documents/DA_L7/L7_term2/visualization/wk6_in_class_AAPL.csv")
dataset
```

```python
dataset.info()
```

```python
# we need to convert the Date column into dates ( above shows as an object )
# we need to convert to date and time format

dataset["Date"]=pd.to_datetime(dataset["Date"],dayfirst=True)
```

```python
dataset.info()
```

```python
# we set the column "Date" as the index

dataset.set_index("Date",inplace=True)
```

```python
dataset.info()
```

```python
# use matplotlib to create the plot

plt.plot(dataset.index,dataset["Close"],label="Close Price",color="blue")
plt.title("AAPL Stock Market Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
```

```python
# use matplotlib to create the plot
# ADD GRIDS and change the size

plt.figure(figsize=(14,7))
plt.plot(dataset.index,dataset["Close"],label="Close Price",color="blue")
plt.title("AAPL Stock Market Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.grid(True)
```

# Now we plot it with seaborn

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=dataset,x=dataset.index,y="Close",label="Close Price", color="red")
sns.lineplot(data=dataset,x=dataset.index,y="Open",label="Open Price",color="blue")

#seaborn is build ontop of matplotlib

plt.title("AAPL Stock Market Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
```
