```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
sns.set();
```

```python
dataset = pd.read_csv('/Users/janet.xuishi/Documents/DA_L7/L7_term2/wk5_Body_fat.csv')
dataset.head()
```

```python
# we want to check relation between chest and abdomen fat
# x represent independant variable, y represent dependent variable

sns.relplot(x='Abdomen', y='Chest',data=dataset)
```

```python
# we can differenciate between the data obtained different days of the week

sns.relplot(x='Abdomen', y='Chest',hue='Day',data=dataset)
```

```python
# we can differenciate between the data obtained according to gendre

sns.relplot(x="Abdomen",y ="Chest",hue ="Gender",data=dataset)
```

```python
# we create a boxplot

sns.catplot(x='Abdomen',y='Gender',kind='box', data=dataset)
```

```python
# we compare the abdomen fat per gender using a barchat

sns.barplot(x='Abdomen',y='Gender',data=dataset)
```

```python
# we compare the abdomen fat per gender using a barchat

sns.barplot(y='Abdomen',x='Gender',data=dataset)
```
