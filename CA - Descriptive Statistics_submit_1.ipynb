```python
import pandas as pd
import statistics as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Descriptive Statistics 1 - Water Consumption in Australia in 2021
# Create a DataFrame to represent Water Consumption in AUS in 2021 data among 6 sectors, total 100%
water_con_data = {
    'Sector': ['R(house)', 'R(apartments)', 'Business', 'Government', 'Industry', 'Other'],
    'AUS Water Consumption (%)': [57, 13, 10, 6, 11, 3]
}

# Convert data to a DataFrame
df = pd.DataFrame(water_con_data)
water_con_df = pd.DataFrame(water_con_data)
```

```python
# Calculate mean, median, mode, max, min,range and std for AUS Water Consumption
water_con_mean = water_con_df['AUS Water Consumption (%)'].mean()
water_con_median = water_con_df['AUS Water Consumption (%)'].median()
water_con_mode = water_con_df['AUS Water Consumption (%)'].mode()
water_con_max = water_con_df['AUS Water Consumption (%)'].max()
water_con_min = water_con_df['AUS Water Consumption (%)'].min()
water_con_range = water_con_max - water_con_min
#water_con_std = np.std(water_con_df['AUS Water Consumption (%)'],ddof=1)
water_con_std = np.std(water_con_df['AUS Water Consumption (%)'])
```

```python
print(f"Mean: {water_con_mean}%")
print(f"Median: {water_con_median}%")
print(f"Mode: \n{water_con_mode}%")

print(f"Max: {water_con_max}%")
print(f"Min: {water_con_min}%")
print(f"Range: {water_con_range}%")
print(f"Standard Deviations: {water_con_std}%")
```

```python
# Descriptive Statistics 1 - Residential Water Usage in Australia in 2021
# Create a DataFrame to represent Residential Water Usage in Australia in 2021 among 6 residential areas, total 100%

residential_data = {
    'Usage': ['Bathroom', 'Toilet', 'Washing Clothes', 'Kitchen', 'Garden'],
    'AUS Residential Water Consumption (%)': [28, 15, 20, 10, 27]
}
# Convert data to a DataFrame
df = pd.DataFrame(residential_data)
residential_df = pd.DataFrame(residential_data)
```

```python
# Calculate mean, median, mode, max, min,range and std for AUS Residential Water Consumption
residential_mean = residential_df['AUS Residential Water Consumption (%)'].mean()
residential_median = residential_df['AUS Residential Water Consumption (%)'].median()
residential_mode = residential_df['AUS Residential Water Consumption (%)'].mode()

residential_max = residential_df['AUS Residential Water Consumption (%)'].max()
residential_min = residential_df['AUS Residential Water Consumption (%)'].min()
residential_range = residential_max - residential_min
residential_std = np.std(residential_df['AUS Residential Water Consumption (%)'])
```

```python
print(f"Mean: {residential_mean}%")
print(f"Median: {residential_median}%")
print(f"Mode: \n{residential_mode}%")

print(f"Max: {residential_max}%")
print(f"Min: {residential_min}%")
print(f"Range: {residential_range}%")
print(f"Standard Deviations: {residential_std}%")
```

```python
# Sort the DataFrame by Residential Water Usage in Australia in 2021 (%) in descending order
residential_df_sorted = residential_df.sort_values(by='AUS Residential Water Consumption (%)', ascending=False)

# Data for the sorted histogram
usage_labels_sorted = residential_df_sorted['Usage']
usage_percentages_sorted = residential_df_sorted['AUS Residential Water Consumption (%)']

# Create a sorted histogram
plt.figure(figsize=(10, 6))
plt.bar(usage_labels_sorted, usage_percentages_sorted,)
plt.xlabel('AUS Residential Water Consumption Areas')
plt.ylabel('Percentage')
plt.title('Distribution of AUS Residential Water Consumption Australia in 2021 (Ranked)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
