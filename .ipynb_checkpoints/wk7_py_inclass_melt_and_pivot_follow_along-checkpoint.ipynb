```python
temperatures = pd.DataFrame({
'city': ['New York', 'London', 'Paris', 'Berlin', 'Amsterdam'],
'day1': [23, 25, 27, 26, 24],
'day2': [22, 21, 25, 26, 23],
'day3': [26, 25, 24, 27, 23],
'day4': [23, 21, 22, 26, 27],
'day5': [27, 26, 27, 24, 28]
})
temperatures
```

```python
# 'id_vars' (any columns we want to keep, don't want to melt)
temperatures.melt(id_vars = ["city"],
                 var_name="date",
                 value_name="temp")
```
