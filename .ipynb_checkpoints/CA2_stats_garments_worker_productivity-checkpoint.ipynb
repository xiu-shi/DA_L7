# We need to find out if the population MEAN is different from 6300 before implementing the new machines.

```python
def calculate_mean_overtime(df):
    mean_overtime_before = round(df['over_time'].mean(),0)
    return mean_overtime_before

before_mean_overtime = calculate_mean_overtime(df)
print(f"Average overtime before: {before_mean_overtime}")
```

# Given after average over time is 6300 minutes

```python
after_mean_overtime = 6300.00
print(f"Average overtime after: {after_mean_overtime}")
```

# Calculate the difference between the two means

```python
diff_mean = before_mean_overtime - after_mean_overtime
print(f"The before and after mean is: {diff_mean}")
```
